```python
import os

dataset_dir = '/content/welding_dataset'
for folder in os.listdir(dataset_dir):
    folder_path = os.path.join(dataset_dir, folder)
    if os.path.isdir(folder_path):
        print(f"Kelas {folder}: {len(os.listdir(folder_path))} gambar")
```

```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator

# Definisikan augmentasi untuk data Training
train_datagen = ImageDataGenerator(
    rescale=1./255,            # Langkah 2: Normalisasi piksel ke 0-1
    rotation_range=15,         # Mengizinkan rotasi acak hingga 15 derajat
    width_shift_range=0.1,     # Pergeseran horizontal acak
    height_shift_range=0.1,    # Pergeseran vertikal acak
    horizontal_flip=True,      # Membalik gambar secara horizontal (kiri-kanan)
    vertical_flip=False,       # Jangan dibalik atas-bawah (agar logis dengan posisi las asli)
    validation_split=0.2       # Aturan 80% Train, 20% Validation/Test
)

# Generator untuk memuat data dari folder langsung ke model
train_generator = train_datagen.flow_from_directory(
    '/content/welding_dataset',
    target_size=(224, 224),    # Langkah 1: Ubah semua resolusi menjadi 224x224
    batch_size=32,
    class_mode='categorical',   # Karena ada 6 kelas (Multi-class)
    subset='training'          # Alokasi untuk data training (80%)
)

validation_generator = train_datagen.flow_from_directory(
    '/content/welding_dataset',
    target_size=(224, 224),
    batch_size=32,
    class_mode='categorical',
    subset='validation'        # Alokasi untuk data validasi/uji (20%)
)
```
